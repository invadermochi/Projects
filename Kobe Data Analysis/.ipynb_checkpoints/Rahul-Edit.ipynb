### __Data Cleaning and Data Wrangling__

```python
site = "https://www.basketball-reference.com/players/b/bryanko01.html"
page = requests.get(site)
soup = BeautifulSoup(page.content, "html.parser")
```

```python
table = soup.find_all(href=re.compile("/players/b/bryanko01/gamelog/"))
links = []
for html in table:
    links.append("https://www.basketball-reference.com/" + html['href'])
    
#Truncating the list as duplicate links were listed
del links[20:]

#Retrieving every game Kobe has played based on the links previously retrieved
year_links = []

for year in links:
    year_links.append(year)
```

```python
def gatherdataframes(links):
    
    dataframes = []
    for i in links:
        df = pd.read_html(i)
        dataframe = df[7]
        data = dataframe[dataframe['G'].notna()]
        data = data[data.G != 'G']
        data = data.reset_index(drop=True)
        dataframes.append(data) 
    return dataframes
```

```python
year_links
```

```python
list_of = gatherdataframes(year_links)
df = pd.concat(list_of, ignore_index = True, sort=False)
dataframe = df.drop(columns=['Rk', 'G', 'Date', 'Age', 'Tm', 'GS', 'DRB', 'TRB','AST', 'STL',
                             'BLK', 'PF', 'TOV', 'GmSc', '+/-', 'ORB', 'Unnamed: 5'])
dataframe = dataframe.rename(columns={'Unnamed: 7': 'Outcome'})
```

```python
def removeExtra(string):
    return string[0]

dataframe["W/L"] = dataframe["Outcome"].apply(removeExtra)
```

```python
dataframe.head(5)
```

```python
def convert_to_seconds(string):
    indexN = string.index(':')
    
    minute = string[0:indexN]
    second = string[indexN+1:]
    
    total = int(minute)*60 + int(second)
    return total

dataframe["SP"] = dataframe["MP"].apply(convert_to_seconds)
```

```python
dataframe.head()
```

```python
def get_average(string, seconds):
        
    averages = []
    
    for index, value in string.items():
        points = int(value)
        playtime = seconds[index]
        average = points/playtime
        average = average * 60
        averages.append(average)
        
    series = pd.Series(averages)
    return series

dataframe['Kobe_Average'] = get_average(dataframe['PTS'], dataframe['SP'])
```

```python
dataframe.head()
```

```python
dataframe.describe()
```

```python
def transform_wins(string):
    if string == 'W':
        return 1
    else:
        return 0
    
dataframe["W/L_Transform"] = dataframe["W/L"].apply(transform_wins)
```

```python
dataframe.head()
```

```python
#dataframe = dataframe.sort_values(by='W/L_Transform', ascending=True)
dataframe.head()
```

```python
dataframe = dataframe.reset_index(drop=True)
dataframe = dataframe.reset_index()
def game_count(counter):
    return counter+1

dataframe["Game_Num"] = dataframe['index'].apply(game_count)
dataframe = dataframe.drop(columns = ['index'])
dataframe.head()
```

```python
def Loss_Performance(avg, win_loss):
    perf = []
    
    for index, value in win_loss.items():
        if value == 0:
            perf.append(avg[index])
    temp = np.array(perf)
    df = pd.DataFrame(temp)
    return df

def Win_Performance(avg, win_loss):
    perf = []
    
    for index, value in win_loss.items():
        if value == 1:
            perf.append(avg[index])
    temp = np.array(perf)
    df = pd.DataFrame(temp)
    return df

df_losses = Loss_Performance(dataframe['Kobe_Average'], dataframe['W/L_Transform'])
df_wins = Win_Performance(dataframe['Kobe_Average'], dataframe['W/L_Transform'])
```

```python
#dataframe.plot(kind='line',x='Game_Num', y='Loss_Perf',color='red')
#dataframe.plot(kind='line',x='Game_Num', y='Win_Perf', color='blue')
#plt.show()
df_losses.columns = ['Kobe_Average']
df_wins.columns = ['Kobe_Average']
```

```python
df_losses.plot(kind='line', y='Kobe_Average',color='red')
df_wins.plot(kind='line', y='Kobe_Average',color='blue')
```
